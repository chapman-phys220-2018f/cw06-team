## Plotting the Trajectories of a Generic Particle, Proton, and Electron

```python
particle = elementary.Particle(0, 0, 1)
electron = elementary.Electron(0, 0, 2)
proton = elementary.Proton(0, 0, 3)

particle.impulse(0, 0, particle.mass*10)
electron.impulse(0, 0, electron.mass*10)
proton.impulse(0, 0, proton.mass*10)

# Creating a figure

# Setting the figure size (width, height)
figure = plt.figure(figsize=(8,6))

# Creating plot axes
axes = plt.axes()

#Creating x-axis values
time_pts = [dt for dt in numpy.arange(0, 4, 1e-2)]

particle_z_pos = []
electron_z_pos = []
proton_z_pos = []

#Setting initial positions
particle_z_pos.append(1)
electron_z_pos.append(2)
proton_z_pos.append(3)


#Creating y-axis values
for i in range(1, len(time_pts)):
    particle.impulse(0, 0, -particle.mass*(9.81)*(time_pts[i]-time_pts[i-1]))
    electron.impulse(0, 0, -electron.mass*(9.81)*(time_pts[i]-time_pts[i-1]))
    proton.impulse(0, 0, -proton.mass*(9.81)*(time_pts[i]-time_pts[i-1]))

    particle.move((time_pts[i]-time_pts[i-1]))
    electron.move((time_pts[i]-time_pts[i-1]))
    proton.move((time_pts[i]-time_pts[i-1]))

    particle_z_pos.append(particle.position[2])
    electron_z_pos.append(electron.position[2])
    proton_z_pos.append(proton.position[2])

#Creating each line on the graph
axes.plot(time_pts, particle_z_pos, label="Particle")
axes.plot(time_pts, electron_z_pos, color="Red", label="Electron")
axes.plot(time_pts, proton_z_pos, color="Blue", label="Proton")
axes.legend()

#Displaying graph
plt.show()
```
